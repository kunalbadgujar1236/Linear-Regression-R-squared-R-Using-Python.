$\textcolor{Yellow}{\Huge\textbf{R-Squared}}$


#### *What is R-Squared?*

- R-squared tells us *how well a line fits the data in a linear regression model.* 
- It shows how much of the changes in the "output" (dependent variable, Y) can be explained by the "input" (independent variable, X).

#### *Example of R-squared (R²) in Simple Terms*
Let’s say you want to predict exam scores based on study hours. You collect data from 5 students:


| Study Hours (X) | Exam Score (Y) |
|-----------------|---------------|
| 2              | 50            |
| 4              | 60            |
| 6              | 70            |
| 8              | 80            |
| 10             | 90            |

A straight line is drawn to fit the data. The equation might look like:

$$ Score = 5 × Study Hours + 40  $$

*What Does This Mean?*

This is a linear equation that helps predict exam scores based on study hours.

- Study Hours → The number of hours a student studies.

- Score → The predicted exam score.

- 5 → This is the slope (also called the coefficient). It means for every extra hour of study, the score increases by 5 points.

- 40 → This is the intercept (also called the constant). It means if a student studies 0 hours, the predicted score is 40.

    | Study Hours (X) | Predicted Score (Y) | Calculation |
    |-----------------|---------------------|-------------|
    | 0              | 40                  | 5(0) + 40 = 40|
    | 2              | 50                  | 5(2) + 40 = 50|
    | 4              | 60                  | 5(4) + 40 = 60|
    | 6              | 70                  | 5(6) + 40 = 70|
    | 8              | 80                  | 5(8) + 40 = 80|
    | 10             | 90                  | 5(10) + 40 = 90|

So, *if you study 6 hours, the model predicts your score will be 70.*

Try running this via Python code:


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Example data (Study Hours vs Exam Scores with some random variation)
np.random.seed(42)  # To get the same random values every time
study_hours = np.array([2, 4, 6, 8, 10]).reshape(-1, 1)  # X (independent variable)
exam_scores = np.array([50, 62, 68, 83, 88]) + np.random.randint(-5, 5, size=5)  # Y (dependent variable with noise)

# Create and train the Linear Regression model
model = LinearRegression()
model.fit(study_hours, exam_scores)

# Predict scores
predicted_scores = model.predict(study_hours)

# Calculate R-squared (R²)
r2 = r2_score(exam_scores, predicted_scores)

# Print R² value
print(f"R-squared (R²): {r2:.2f}")

# Plot the data
plt.scatter(study_hours, exam_scores, color='blue', label="Actual Scores")
plt.plot(study_hours, predicted_scores, color='red', linestyle="--", label="Regression Line")
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.title(f"Linear Regression (R² = {r2:.2f})")
plt.legend()
plt.show()
```
